```python
import pandas as pd
import numpy as np
import os
```

```python
def ab():
    file_paths = []
    data_dir_path = 'Data/'
    elemnts = os.listdir(data_dir_path)
    for item in elemnts:
        if item.endswith('.csv'):
            file_path = os.path.join('Data/',item)
            file_paths.append(file_path)
    return file_paths    
```

```python
file_path = ab()
```

```python
df = pd.read_csv(file_path[0])
df2 = pd.read_csv(file_path[1])
```
